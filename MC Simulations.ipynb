```python
import numpy as np
from MC_Simulations_Chain import MCChain
from AGZChainEnv import Chain
from Chain_Env import ChainEnv
```

```python
seq = "PPPHHPHPHPPHHHP"
env = Chain(seq, max_len = 15, grid_len = 31, max_moves = 10000)
mc_sim = MCChain(env)
state = env.make_state()
```

```python
score, info = mc_sim.sim(state, 1)
```

```python
print(score)
print(info)
```

```python
visual_env = ChainEnv("PPPHHPHPHPPHHHP", 15, 31)
visual_env.render()

for a in info['Actions']:
    visual_env.step(a)
    visual_env.render()
```
